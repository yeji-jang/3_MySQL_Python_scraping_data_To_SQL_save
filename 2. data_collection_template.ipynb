## [#1 : 마블 캐릭터]

### 1-1.   데이터 정의

- 마블 캐릭터 데이터는 
- ['id', 'name', 'description'] 3개의 columns와 1,562개의 rows로 구성되어 있습니다.

### 1-2. 수집 방법

* MARVEL 공식 홈페이지의 오픈 API 활용
* API 파라미터 : (필수) serviceKey=인증키, (선택) hash key=식별값, limit = 100, offset = 0~n번 반복   
* requests.get한 URL을 [data]-[results] 값만 가져와 json으로 변환
* for문을 사용하여, [results]에 들어있는 각각의 key값을 빈{ }에 담고 최종결과는 [ ]에 append 했습니다.
* 최종결과 리스트를 pandas의 DataFrame으로 변환하고 그것을 csv 형식으로 저장하였습니다.

### 1-3.   수집 경로

* 마블 캐릭터 url
- URL : (https://developer.marvel.com/docs#!/public/getCreatorCollection_get_0)  

### 1-4.   데이터의 논리적 구조

* ID : 캐릭터의 고유값을 나타내는 정수 값입니다.
* name : 캐릭터의 이름을 나타낸 str 객체입니다.
* description : 짧은 약력 또는 캐릭터에 대해 설명한 str 객체입니다. 

### 1-5.   데이터의 물리적 저장 구조
